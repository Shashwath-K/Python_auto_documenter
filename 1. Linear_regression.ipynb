``` text 
Linear regression is a statistical technique used to find the relationship between variables. In an ML context, linear regression finds the relationship between features and a label.

In other words

Linear regression attempts to model the relationship between two variables by fitting a linear equation to observed data. One variable is considered to be an explanatory variable, and the other is considered to be a dependent variable. For example, a modeler might want to relate the weights of individuals to their heights using a linear regression model.
```

> For example, suppose we want to predict a car's fuel efficiency in miles per gallon based on how heavy the car is, and we have the following dataset:

| Pounds in 1000s (feature) | Miles per gallon (label) |
|---------------------------|--------------------------|
| 3.5	| 18 |
| 3.69	| 15 |
| 3.44	| 18 |
| 3.43	| 16 |
| 4.34	| 15 |
| 4.42	| 14 |
| 2.37	| 24 |

*Following program demonstrates the values mentioned in the above table and gives a visual representation using scatter plot.*

````python
import matplotlib.pyplot as plt

x = [3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37]
y = [18, 15, 18, 16, 15, 14, 24]

plt.scatter(x,y)
plt.xlabel("Pounds in 1000s (Feature)")
plt.ylabel("Miles per Gallon (Label)")
plt.show()
````

### #2. Using pyplot and scipy (built-in libs) to perform linear regression using scatter-plot and stats [from scipy].

````python
import matplotlib.pyplot as plt
from scipy import stats

x = [3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37]
y = [18, 15, 18, 16, 15, 14, 24]

slope, intercept, r_value, p_value, std_error = stats.linregress(x,y)

def predict(x):
    return slope * x + intercept

predict_value = list(map(predict, x))

plt.scatter(x,predict_value)
plt.plot(x, predict_value, color='green')
plt.xlabel("Pounds in 1000s (Feature)")
plt.ylabel("Miles per Gallon (Label)")
plt.show()
````

### #3. Using RAW approach with the step-by-step iteration of the linear regression methods

````python
import matplotlib.pyplot as plt

class Solution:
    def calculate_equation(self, x, y):
        n = len(x)
        n1 = len(y)
        if n != n1:
            return "The number of independent and dependent variables must be the same"
        # Calculate the x^2, y^2 and x*y for each value and store the items in a list
        x_squared = [i**2 for i in x]
        y_squared = [i**2 for i in y]
        xy = [x[i]*y[i] for i in range(n)]

        # Calculate the sum of x, y, x^2, y^2 and x*y
        sum_x = sum(x)
        sum_y = sum(y)
        sum_x_squared = sum(x_squared)
        sum_y_squared = sum(y_squared)
        sum_xy = sum(xy)

        # Calculate the slope (m) and intercept (b) using the formulas
        m = (n*sum_xy - sum_x*sum_y) / (n*sum_x_squared - sum_x**2)
        b = (sum_y - m*sum_x) / n

        # Return the slope and intercept as a values
        return m, b
    
    def linear_regression(self, x, y):
        m, b = self.calculate_equation(x, y)
        
        plt.scatter(x,y)
        plt.plot(x, [m*i + b for i in x], color='green')
        plt.xlabel("Independent Variable (x)")
        plt.ylabel("Dependent Variable (y)")
        plt.show()

if __name__ == "__main__":

    # Input the variables 
    # Decide which is the Independent variable (x) and which is the Dependent variable (y)
    x = [2,4,2,5,4,0,3,5,1,3]
    y = [137, 70, 184, 0, 35, 297, 122 , 1, 253, 150]

    solution = Solution()
    result = solution.linear_regression(x,y)
````
